```python
import pandas as pd
import numpy as np
import datetime
from datetime import datetime




adress1 = r"Real_Estate_Sales_2001-2020_GL 1.parquet"
adress2 = r"Real_Estate_Sales_2001-2020_GL 2.parquet"
df_1 = pd.read_parquet(adress1)
df_2 = pd.read_parquet(adress2)

df = pd.concat([df_1,df_2])
df
```

```python
#Comparison of houses sold in every town in 2015 and 2019

values_2015 = {"Serial Number" : "Total sales 2015"}
sales_2015 = df.loc[df["Date Recorded"].dt.year==2015].groupby("Town")[["Serial Number"]].count().reset_index().rename(columns = values_2015)

values_2016 = {"Serial Number" : "Total sales 2016"}
sales_2016 = df.loc[df["Date Recorded"].dt.year==2016].groupby("Town")[["Serial Number"]].count().reset_index().rename(columns = values_2016)


total_sale = sales_2015.merge(sales_2016, how= "left", left_on="Town", right_on = "Town")

total_sales = total_sale.copy()

total_sales["Total"] = total_sales["Total sales 2015"]-total_sales["Total sales 2016"]
total_sales
```

```python
#Comparison of houses average sold price in 2012 and 2016 and ordered by price diference


values_2015 = {"Sale Amount" : "Avg price 2015"}
price_2015 = df.loc[df["Date Recorded"].dt.year==2015].groupby("Town")[["Sale Amount"]].mean().reset_index().rename(columns = values_2015)


values_2016 = {"Sale Amount" : "Avg price 2016"}
price_2016 = df.loc[df["Date Recorded"].dt.year==2016].groupby("Town")[["Sale Amount"]].mean().reset_index().rename(columns = values_2016)



total_price = price_2015.merge(price_2016, how= "left", left_on="Town", right_on = "Town")

avg_sales = total_price.copy()

def calculo(avg_sales):
    valor = ((avg_sales["Avg price 2015"] / avg_sales["Avg price 2016"]))*100
    return 100-round(valor,2)
    
avg_sales["% Diference"] = avg_sales.apply(calculo, axis = 1)
# avg_sales
avg_sales.sort_values(by = "% Diference", ascending = False)
```

```python
#Create a column that only is "CHECK" if the sale is between 2011 and 2013, New London town and the assessed price is less  of 500000
df["Check"] = np.where(

    ((df["Date Recorded"].dt.year>= 2011) &
    (df["Date Recorded"].dt.year<= 2013)) &
    (df["Town"]=="New London") & 
    (df["Assessed Value"]>500000), "Check","-")

df.loc[df["Check"] == "Check"]
```
